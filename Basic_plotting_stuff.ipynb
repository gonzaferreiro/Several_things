```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
from scipy import spatial
import re

plt.style.use('fivethirtyeight')

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
weight = [5,4,8,2,6,2]
month = ['jan','feb','mar','apr','may','jun']

animal_type = ['dog','cat','cat','dog','dog','dog']
hue = ['blue','red','red','blue','blue','blue']

average_animal_weight = [4,5,7,5,3,2]

size = [2,3,5,1,4,1]
```

```python
g = sns.scatterplot(x=month, y=weight, hue=hue, legend=False, label='animals measured')
g.axvline(2,c='r')

ax2 = g.twinx()
sns.lineplot(x=month, y=average_animal_weight, ax=ax2, c='y', label='average weight')

g.legend()
plt.show()
```
